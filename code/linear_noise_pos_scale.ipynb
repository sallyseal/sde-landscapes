```julia
##################################################
# ANALYTICAL STEADY STATE PROBABILITY DISTRIBUTION
##################################################
```

```julia
α=0.04
ssd_1(x) = exp((2/α^2)*((-a1*(x-1.5)^4)/4 + (b1*(x-1.5)^3)/3 + (c1*(x-1.5)^2)/2 + d1*(x-1.5)))
p1 = plot(ssd_1,-1,4)

α=0.1
ssd_2(x) = exp((2/α^2)*((-a1*(x-1.5)^4)/4 + (b1*(x-1.5)^3)/3 + (c1*(x-1.5)^2)/2 + d1*(x-1.5)))
p2 = plot(ssd_2,-1,4)

α=0.5
ssd_3(x) = exp((2/α^2)*((-a1*(x-1.5)^4)/4 + (b1*(x-1.5)^3)/3 + (c1*(x-1.5)^2)/2 + d1*(x-1.5)))
p3 = plot(ssd_3,-1,4)

α=1.0
ssd_4(x) = exp((2/α^2)*((-a1*(x-1.5)^4)/4 + (b1*(x-1.5)^3)/3 + (c1*(x-1.5)^2)/2 + d1*(x-1.5)))
p4 = plot(ssd_4,-1,4)

final_ssd = plot(p1,p2,p3,p4,layout=(4,1),size=(700,900))
```

```julia
##################################################
# ADJUSTING NOISE LEVELS AND EFFECT ON LANDSCAPE
##################################################
```

```julia
α=0.04
land_1(x) = -(2/α^2)*((-a1*(x-1.5)^4)/4 + (b1*(x-1.5)^3)/3 + (c1*(x-1.5)^2)/2 + d1*(x-1.5))
pl1 = plot(land_1,-1,4,label="Landscape: noise=$α", xlabel=("x"),ylabel=("U(x)"))

α=0.1
land_2(x) = -(2/α^2)*((-a1*(x-1.5)^4)/4 + (b1*(x-1.5)^3)/3 + (c1*(x-1.5)^2)/2 + d1*(x-1.5))
pl2 = plot(land_2,-1,4,label="Landscape: noise=$α", xlabel=("x"),ylabel=("U(x)"))

α=0.5
land_3(x) = -(2/α^2)*((-a1*(x-1.5)^4)/4 + (b1*(x-1.5)^3)/3 + (c1*(x-1.5)^2)/2 + d1*(x-1.5))
pl3 = plot(land_3,-1,4,label="Landscape: noise=$α", xlabel=("x"),ylabel=("U(x)"))

α=1.0
land_4(x) = -(2/α^2)*((-a1*(x-1.5)^4)/4 + (b1*(x-1.5)^3)/3 + (c1*(x-1.5)^2)/2 + d1*(x-1.5))
pl4 = plot(land_4,-1,4,label="Landscape: noise=$α", xlabel=("x"),ylabel=("U(x)"))

final_land = plot(pl1,pl2,pl3,pl4,layout=(4,1),size=(700,900))
```
